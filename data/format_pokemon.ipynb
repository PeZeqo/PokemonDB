```python
import pandas as pd
import numpy as np
```

```python
pkm = pd.read_csv('pokemon_table.csv', engine='python')
```

```python
pkm.head()
```

```python
pkm['Type1ID'] = np.nan
```

```python
pkm['Type2ID'] = np.nan
```

```python
pkm.head()
```

```python
count = 0
for row in pkm.iterrows():
    if row[1][2] == 'Bug':
        pkm.at[count, 'Type1ID'] = 1
    if row[1][3] == 'Bug':
        pkm.at[count, 'Type2ID'] = 1
    if row[1][2] == 'Dark':
        pkm.at[count, 'Type1ID'] = 2
    if row[1][3] == 'Dark':
        pkm.at[count, 'Type2ID'] = 2
    if row[1][2] == 'Dragon':
        pkm.at[count, 'Type1ID'] = 3
    if row[1][3] == 'Dragon':
        pkm.at[count, 'Type2ID'] = 3
    if row[1][2] == 'Electric':
        pkm.at[count, 'Type1ID'] = 4
    if row[1][3] == 'Electric':
        pkm.at[count, 'Type2ID'] = 4
    if row[1][2] == 'Fairy':
        pkm.at[count, 'Type1ID'] = 5
    if row[1][3] == 'Fairy':
        pkm.at[count, 'Type2ID'] = 5
    if row[1][2] == 'Fighting':
        pkm.at[count, 'Type1ID'] = 6
    if row[1][3] == 'Fighting':
        pkm.at[count, 'Type2ID'] = 6
    if row[1][2] == 'Fire':
        pkm.at[count, 'Type1ID'] = 7
    if row[1][3] == 'Fire':
        pkm.at[count, 'Type2ID'] = 7
    if row[1][2] == 'Flying':
        pkm.at[count, 'Type1ID'] = 8
    if row[1][3] == 'Flying':
        pkm.at[count, 'Type2ID'] = 8
    if row[1][2] == 'Ground':
        pkm.at[count, 'Type1ID'] = 9
    if row[1][3] == 'Ground':
        pkm.at[count, 'Type2ID'] = 9
    if row[1][2] == 'Grass':
        pkm.at[count, 'Type1ID'] = 10
    if row[1][3] == 'Grass':
        pkm.at[count, 'Type2ID'] = 10
    if row[1][2] == 'Ghost':
        pkm.at[count, 'Type1ID'] = 11
    if row[1][3] == 'Ghost':
        pkm.at[count, 'Type2ID'] = 11
    if row[1][2] == 'Ice':
        pkm.at[count, 'Type1ID'] = 12
    if row[1][3] == 'Ice':
        pkm.at[count, 'Type2ID'] = 12
    if row[1][2] == 'Normal':
        pkm.at[count, 'Type1ID'] = 13
    if row[1][3] == 'Normal':
        pkm.at[count, 'Type2ID'] = 13
    if row[1][2] == 'Poison':
        pkm.at[count, 'Type1ID'] = 14
    if row[1][3] == 'Poison':
        pkm.at[count, 'Type2ID'] = 14
    if row[1][2] == 'Water':
        pkm.at[count, 'Type1ID'] = 15
    if row[1][3] == 'Water':
        pkm.at[count, 'Type2ID'] = 15
    if row[1][2] == 'Psychic':
        pkm.at[count, 'Type1ID'] = 16
    if row[1][3] == 'Psychic':
        pkm.at[count, 'Type2ID'] = 16
    if row[1][2] == 'Rock':
        pkm.at[count, 'Type1ID'] = 17
    if row[1][3] == 'Rock':
        pkm.at[count, 'Type2ID'] = 17
    if row[1][2] == 'Steel':
        pkm.at[count, 'Type1ID'] = 18
    if row[1][3] == 'Steel':
        pkm.at[count, 'Type2ID'] = 18
    count += 1
```
